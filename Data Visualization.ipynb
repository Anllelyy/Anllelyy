```python
# Dataframe
Heart_Disease = df['Heart_Disease'].value_counts()

# Data for the pie chart
labels = Heart_Disease.index
sizes = Heart_Disease.values
colors=[ 'crimson', 'orange', 'crimson','gold', 'limegreen', 'darkturquoise', 'deepskyblue','mediumblue','rebeccapurple','darkorchid','deeppink','hotpink']

# Create the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=4)

# Legend labels
legend_labels = ['0 - No Heart Disease', '1 - Yes, have Heart Disease']  # Replace with actual names
plt.legend(legend_labels, title='Category', bbox_to_anchor=(1, 0.5), loc='center left')

# Add title
plt.title('Distribution of Heart_Disease')

# Show the plot
plt.show()
```

```python
from scipy.stats import gamma, kstest
import numpy as np

# Supongamos que 'data' es tu conjunto de datos
data = np.array(df['Alcohol_Consumption'])

# ParÃ¡metros estimados de la distribuciÃ³n gamma
alpha, loc, beta = gamma.fit(data)

# Genera datos de la distribuciÃ³n gamma ajustada
gamma_fitted = gamma.rvs(alpha, loc=loc, scale=beta, size=len(data))

# Prueba de Kolmogorov-Smirnov
ks_statistic, ks_p_value = kstest(data, 'gamma', args=(alpha, loc, beta))

print(f"EstadÃ­stico de prueba KS: {ks_statistic}")
print(f"P-valor: {ks_p_value}")

# Puedes usar ks_p_value para tomar decisiones sobre la hipÃ³tesis nula
# Por ejemplo, si ks_p_value < 0.05, podrÃ­as rechazar la hipÃ³tesis nula de que los datos siguen una distribuciÃ³n gamma.
```

```python
from scipy.stats import gamma, kstest

data = df['Alcohol_Consumption']  # You can replace this data with your own

# Perform the Kolmogorov-Smirnov test
ks_statistic, p_value = kstest(data, 'gamma', args=(2,))  # 'gamma' indicates the gamma distribution with parameter 'a=2'

# Print results
print(f'KS test statistic: {ks_statistic}')
print(f'P-value: {p_value}')

# Make a decision based on the p-value (using a significance level of 0.05)
significance_level = 0.05
if p_value > significance_level:
    print('Fail to reject the null hypothesis. The data follows a gamma distribution.')
else:
    print('Reject the null hypothesis. The data does not follow a gamma distribution.')

# Visualize the data and the fitted gamma distribution for illustrative purposes
plt.hist(data, bins=30, density=True, alpha=0.5, color='blue', label='Observed data')
x = np.linspace(0, max(data), 100)
plt.plot(x, gamma.pdf(x, a=2), 'r-', lw=2, label='Fitted gamma distribution (a=2)')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Goodness of Fit Test to Gamma Distribution')
plt.legend()
plt.show()
```
